```python
import sys
import time

def convert_numbers(file_path):
    try:
        with open(file_path, 'r') as file:
            numbers = [int(line.strip()) for line in file]

        binary_numbers = [bin(num)[2:] for num in numbers]
        hex_numbers = [hex(num)[2:].upper() for num in numbers]

        with open('ConversionResults.txt', 'w') as result_file:
            result_file.write("NUMBER\tTC1\tBIN\tHEX\n")
            for i, num in enumerate(numbers, start=1):
                tc1 = str(num)
                result_file.write(f"{i}\t{tc1}\t{binary_numbers[i-1]}\t{hex_numbers[i-1]}\n")

        print("Conversion results:")
        print("NUMBER\tTC1\tBIN\tHEX")
        for i, num in enumerate(numbers, start=1):
            tc1 = str(num)
            print(f"{i}\t{tc1}\t{binary_numbers[i-1]}\t{hex_numbers[i-1]}")

        return True

    except Exception as e:
        print(f"Error: {e}")
        return False

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python convertNumbers.py fileWithData.txt")
    else:
        start_time = time.time()
        success = convert_numbers(sys.argv[1])
        end_time = time.time()

        if success:
            elapsed_time = end_time - start_time
            print(f"\nTime elapsed: {elapsed_time:.4f} seconds")
            with open('ConversionResults.txt', 'a') as result_file:
                result_file.write(f"\nTime elapsed: {elapsed_time:.4f} seconds")
```

```python
convert_numbers("/content/TC2.txt")
```
